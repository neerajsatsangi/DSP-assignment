```python
##  5
import thinkdsp as t
import thinkplot
from IPython.display import Audio

s1 = t.SinSignal(freq = 140, amp = 1.3, offset = 0)
s1_w = s1.make_wave(start = 0, duration = 2, framerate = 48000)
s2 = t.SinSignal(freq = 480, amp = 2, offset = 0)
s2_w = s2.make_wave(start = 0, duration = 2, framerate = 48000)
s3 = t.SinSignal(freq = 720, amp = 1.4, offset = 0)
s3_w = s3.make_wave(start = 0, duration = 2, framerate = 48000)
s4 = t.SinSignal(freq = 240,amp = 1,offset = 0)
s4_w = s4.make_wave(start = 0, duration = 2, framerate = 48000)
s5 = t.SinSignal(freq = 240,amp = 1.2,offset = 0)
s5_w = s5.make_wave(start = 0, duration = 2, framerate = 48000)
s6 = t.SinSignal(freq = 480,amp = 1.5,offset = 0)
s6_w = s6.make_wave(start = 0, duration = 2, framerate = 48000)
s7 = t.SinSignal(freq = 120,amp = 1,offset = 0)
s7_w = s7.make_wave(start = 0, duration = 2, framerate = 48000)

add = s1_w + s2_w + s3_w + s4_w + s5_w + s6_w + s7_w

audio = Audio(data = add.ys, rate = add.framerate)
audio
```
